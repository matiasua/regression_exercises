```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import model_selection
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('titanic.csv', sep=',')
df
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], 1)
df.shape
```

### Replazo los valores de tipo string a entero

    0 - male
    1 - female
    
###### Pregunta: sera esta forma correcta de hacer el remplazo de estos valores?


###### tiene un sentido logico el 0 - male, 1 - female en la dimensionalidad del modelo?

```python
df['Sex']=df['Sex'].replace('male',0)
df['Sex']=df['Sex'].replace('female',1)
df
```

### Replazo los valores de tipo string a entero

    0 - C = Cherbourg
    1 - Q = Queenston
    2 - S = Southapmtpon

```python
df['Embarked']=df['Embarked'].replace('C',0)
df['Embarked']=df['Embarked'].replace('Q',1)
df['Embarked']=df['Embarked'].replace('S',2)
df
```

```python
df.shape
```

```python
# Valores nulos
for feature in df.columns:
    print('Total de valores nulos de', feature, '=', df[feature].isna().sum())
```

### Remuevo los valores nulos

```python
df = df.dropna()
```

```python
df
```

```python
df.describe()
```

```python
print(df.groupby('Pclass').size())
```

```python
df = df.drop(['Sex', 'Fare', 'Parch', 'SibSp'], 1)
df.shape
```

```python
X = np.array(df.drop(['Survived'], 1))
X.shape
```

```python
y = np.array(df['Survived'])
y.shape
```

```python
from sklearn import model_selection
```

```python
validation_size = 0.20
seed = 29
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, y, test_size=validation_size, random_state=seed)
```

```python
model = linear_model.LogisticRegression()
model.fit(X_train, Y_train)
```

```python
name='Model Score'
kfold = model_selection.KFold(n_splits=10, random_state=None)
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)
```

```python
from sklearn.metrics import accuracy_score
```

```python
predictions = model.predict(X_validation)
print(accuracy_score(Y_validation, predictions))
```

```python
print('Score del modelo (train): %.2f' % model.score(X_train, Y_train)) # Return the coefficient of determination  of the prediction.
print('Score del modelo (test): %.6f' % model.score(X_validation, Y_validation))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_validation, predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_validation, predictions))
```
